```python
# Import basics
import sys, os
import numpy as np

from scipy import stats, ndimage, interpolate
import matplotlib as mpl
import matplotlib.pyplot as plt

# Plotting Settings
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = 'CMU Serif'
mpl.rcParams['font.sans-serif'] = 'CMU Sans Serif, DejaVu Sans, Bitstream Vera Sans, Lucida Grande, Verdana, Geneva, Lucid, Arial, Helvetica, Avant Garde, sans-serif'
mpl.rcParams['text.usetex'] = True



Hz2eV = 6.582119514e-16
rbfuf = np.loadtxt('RBFUF.csv', delimiter = ',')
rbfuf[-1] = rbfuf[-2]
```

# Load the Analysis Results

Analysis results are saved for each observation in the directory `../data/Results/` as numpy dictionaries. Each numpy dictionary contains:

    Freqs : The analysis frequency, in GHz
    FD_Limit: The 95% upper limit on the flux density at the analysis frequency, in Jy
    TS_noVetoes: The TS for a central channel excess at the analysis frequency
    TS_wVetoes: The TS for a central channel excess at the analysis frequency after vetoes are applied
    TS_Percentile: The p-value of the TS of the value at the analysis channel or larger determined by Monte Carlo 
    ma: The axion mass corresponding to the analysis frequency
    
    
The `J0720` and `J0806` result dictionaries have a single limit on the axion coupling $g_{a \gamma \gamma}$ in units of $\mathrm{GeV}^{-1}$ contained in `Limit`. All other observations have limits for four neutron star population model and dark matter density profile assumptions in identical units made available in: `ModelI_NFW`, `ModelII_NFW`, `ModelI_Cored`, and `ModelII_Cored`.
    
    

```python
J0720 = np.load('../data/Results/GBT_J0720.npz')['arr_0'].item()
J0806 = np.load('../data/Results/GBT_J0806.npz')['arr_0'].item()

Eff_LBand = np.load('../data/Results/Eff_LBand.npz')['arr_0'].item()
Eff_SBand = np.load('../data/Results/Eff_SBand.npz')['arr_0'].item()
```

# Flux Density Limits

```python
mpl.rcParams['figure.figsize'] = 12, 6
mpl.rcParams['axes.labelsize'] = 24 # default figure size
mpl.rcParams['xtick.labelsize'] = 20
mpl.rcParams['ytick.labelsize'] = 20

f,(ax0,ax1) = plt.subplots(1,2,sharey=True, facecolor='w')


for ax in (ax0, ax1):
    
    ax.plot( Eff_LBand['Freqs'], ndimage.median_filter(Eff_LBand['FD_Limit'], size = 251),
             label = 'Eff GC', lw = 2, c ='black')

    ax.plot(Eff_SBand['Freqs'], ndimage.median_filter( Eff_SBand['FD_Limit'], size = 251),
            lw = 2, c ='black')
    
    max_loc = np.argmax(np.diff(J0720['Freqs']))+1
    
    ax.plot(J0720['Freqs'] [:max_loc], ndimage.median_filter(J0720['FD_Limit'][:max_loc], size = 251),
            label = 'RX J0720', lw = 2, c = 'maroon')
    
    ax.plot(J0720['Freqs'] [max_loc:], ndimage.median_filter(J0720['FD_Limit'][max_loc:], size = 251),
        label = 'RX J0720', lw = 2, c = 'maroon')
    
    
    ax.plot(J0806['Freqs'] [:max_loc], ndimage.median_filter(J0806['FD_Limit'][:max_loc], size = 251),
            lw = 2, c = 'navy')
    
    ax.plot(J0806['Freqs'] [max_loc:], ndimage.median_filter(J0806['FD_Limit'][max_loc:], size = 251),
            lw = 2, c = 'navy')
    
ax1.legend(loc = 'upper right', fontsize = 20)
ax0.set_xlim(1.1, 1.8)
ax1.set_xlim(2.35, 2.75)

ax0.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax0.yaxis.tick_left()
ax0.tick_params(labelright='off')
ax1.yaxis.tick_right()

d = .015 
kwargs = dict(transform=ax0.transAxes, color='k', clip_on=False)
ax0.plot((1-d,1+d), (-d,+d), **kwargs)
ax0.plot((1-d,1+d),(1-d,1+d), **kwargs)

kwargs.update(transform=ax1.transAxes)  # switch to the bottom axes
ax1.plot((-d,+d), (1-d,1+d), **kwargs)
ax1.plot((-d,+d), (-d,+d), **kwargs)
    
    
f.text(0.55, -.04, r'Frequency [GHz]', ha='center', va='center', fontsize = 36)
plt.tight_layout()

ax0.set_yscale('log')
ax0.set_ylabel('Flux Density Limit [Jy]', fontsize = 36)
plt.tight_layout()
plt.show()
```

# Survival Function

```python
# Figure size
mpl.rcParams['figure.figsize'] = 20, 8
mpl.rcParams['axes.labelsize'] = 36 # default figure size
mpl.rcParams['xtick.labelsize'] = 28
mpl.rcParams['ytick.labelsize'] = 28


x_vals = np.logspace(-3, 2, 1000)


#######################################
###   Test Statistics with Vetoes   ###
#######################################

plt.subplot(121)

plt.plot(np.sort(Eff_LBand['TS_wVetoes']), 1 - np.arange(len(Eff_LBand['TS_wVetoes'])) / float(len(Eff_LBand['TS_wVetoes'])),
         lw = 3, label = 'Eff LBand', c = 'cornflowerblue')

plt.plot(np.sort(Eff_SBand['TS_wVetoes']), 1 - np.arange(len(Eff_SBand['TS_wVetoes'])) / float(len(Eff_SBand['TS_wVetoes'])),
         lw = 3, label = 'Eff SBand', c = 'forestgreen')

plt.plot(np.sort(J0720['TS_wVetoes']), 1 - np.arange(len(J0720['TS_wVetoes'])) / float(len(J0720['TS_wVetoes'])),
         lw  =3, label = 'RX J0720', c = 'maroon')
plt.plot(np.sort(J0806['TS_wVetoes']), 1 - np.arange(len(J0806['TS_wVetoes'])) / float(len(J0806['TS_wVetoes'])),
         lw  =3, label = 'RX J0806', c = 'goldenrod')



plt.plot(x_vals, .5*stats.chi2.sf(x_vals, df = 1), c = 'black',
        label = '$\chi^2$ Distribution', lw = 1, ls = ':')

plt.axvline(1e2, c = 'black', ls = '--')

plt.xlim(2e-1,2e2)
plt.ylim(1./len(J0720['TS_wVetoes']),1.1)


plt.yscale('log')
plt.xscale('log')
plt.xlabel('TS')
plt.ylabel('Survival Fraction')

##########################################
###   Test Statistics without Vetoes   ###
##########################################

plt.subplot(122)

plt.plot(np.sort(Eff_LBand['TS_noVetoes']), 1 - np.arange(len(Eff_LBand['TS_noVetoes'])) / float(len(Eff_LBand['TS_noVetoes'])),
         lw = 3, label = 'Eff LBand', c = 'cornflowerblue')

plt.plot(np.sort(Eff_SBand['TS_noVetoes']), 1 - np.arange(len(Eff_SBand['TS_noVetoes'])) / float(len(Eff_SBand['TS_noVetoes'])),
         lw = 3, label = 'Eff SBand', c = 'forestgreen')

plt.plot(np.sort(J0720['TS_noVetoes']), 1 - np.arange(len(J0720['TS_noVetoes'])) / float(len(J0720['TS_noVetoes'])),
         lw  =3, label = 'RX J0720', c = 'maroon')
plt.plot(np.sort(J0806['TS_noVetoes']), 1 - np.arange(len(J0806['TS_noVetoes'])) / float(len(J0806['TS_noVetoes'])),
         lw  =3, label = 'RX J0806', c = 'goldenrod')

x_vals = np.logspace(-3, 2, 1000)

plt.plot(x_vals, .5*stats.chi2.sf(x_vals, df = 1), c = 'black',
        label = '$\chi^2$ Distribution', lw = 1, ls = ':')

plt.axvline(1e2, c = 'black', ls = '--', label = 'Detection Threshold')

plt.xlim(2e-1, 1e6)
plt.ylim(1./len(J0720['TS_wVetoes']),1.1)

plt.legend(ncol = 2, framealpha = 1., fontsize = 20)

plt.yscale('log')
plt.xscale('log')
plt.xlabel('TS')
plt.yticks([])

plt.tight_layout()
plt.subplots_adjust(hspace = .1)
plt.show()
```
